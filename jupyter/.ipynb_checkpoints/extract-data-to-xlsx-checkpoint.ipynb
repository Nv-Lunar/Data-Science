# structuring and exporting data

```python
# Import the relevant modules
import requests
import json
```

```python
# define base URL
base_site = "https://itunes.apple.com/search"
```

```python
# Make a request
res = requests.get(base_site, params = {"term": "the beatles", "country": "us", "limit": 200})
res.status_code
```

```python
# Store the response
info = res.json()
print(json.dumps(info["results"][1], indent = 4))
```

```python
import pandas as pd
```

```python
songs_df = pd.DataFrame(info["results"])
songs_df
```

```python
songs_df.to_csv("songs_info.csv")
```

```python
songs_df.to_excel("info.xlsx")
```
