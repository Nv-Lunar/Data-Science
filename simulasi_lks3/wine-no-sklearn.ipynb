- Fixed Acidity (Asam Tetap):
    Merupakan jumlah asam yang tetap dalam anggur. Asam ini memberikan rasa keasaman pada anggur.
- Volatile Acidity (Asam Volatil):
    Mengukur jumlah asam asetat dalam anggur. Jumlah yang terlalu tinggi dapat memberikan rasa asam yang tidak diinginkan.
- Citric Acid (Asam Sitrat):
    Merupakan jenis asam yang memberikan rasa segar dan menyegarkan pada anggur.
- Residual Sugar (Gula Tersisa):
    Merupakan jumlah gula yang tersisa setelah fermentasi. Memengaruhi tingkat ke-manisanan anggur.
- Chlorides (Klorida):
    Jumlah klorida dalam anggur, yang dapat mempengaruhi rasa dan aroma.
- Free Sulfur Dioxide (Sulfur Dioksida Bebas):
    Jumlah sulfur dioksida yang ada dalam bentuk bebas. Bertindak sebagai pengawet dan antioksidan.
- Total Sulfur Dioxide (Sulfur Dioksida Total):
    Jumlah total sulfur dioksida, termasuk yang berbentuk bebas dan yang terikat dengan senyawa lain.
- Density (Kepadatan):
    Mengukur kepadatan relatif anggur. Dapat terkait dengan kandungan alkohol.
- pH:
    Mengukur tingkat keasaman atau kebasaan anggur. Skala pH berkisar antara 0 (sangat asam) hingga 14 (sangat basa).
- Sulphates (Sulfat):
    Merupakan senyawa sulfur yang dapat memengaruhi aroma dan kestabilan anggur.
- Alcohol (Kandungan Alkohol):
    Persentase alkohol dalam anggur, yang mempengaruhi rasa dan karakteristik umum.

```python
# Linear regresion
def linear_regression_fit(X, y):
    """
    Melakukan fitting model regresi linier.

    Parameters:
    - X: Matriks fitur
    - y: Vektor target

    Returns:
    - coefficients: Koefisien regresi
    """
    X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Menambahkan kolom bias
    coefficients = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
    return coefficients

def linear_regression_predict(X, coefficients):
    """
    Melakukan prediksi menggunakan model regresi linier.

    Parameters:
    - X: Matriks fitur untuk prediksi
    - coefficients: Koefisien regresi

    Returns:
    - predictions: Prediksi target
    """
    X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Menambahkan kolom bias
    predictions = X_b.dot(coefficients)
    return predictions

def r2_score_manual(y_true, y_pred):
    mean_y = np.mean(y_true)
    ss_total = np.sum((y_true - mean_y)**2)
    ss_residual = np.sum((y_true - y_pred)**2)
    r2 = 1 - (ss_residual / ss_total)
    return r2

# def standard_scaler(X):
#     """
#     StandarScaler untuk penskalaan fitur tanpa menggunakan scikit-learn.

#     Parameters:
#     - X: Matriks fitur

#     Returns:
#     - X_scaled: Matriks fitur yang telah di-skala
#     """
#     mean_values = np.mean(X, axis=0)
#     std_values = np.std(X, axis=0)
#     X_scaled = (X - mean_values) / std_values
#     return X_scaled
```

```python
def check_purity(data):
    
    quality_column = data[:, -1]
    unique_classes = np.unique(quality_column)

    if len(unique_classes) == 1:
        return True
    else:
        return False
```

```python
def classify_data(data):
    
    # hanya ambil kolom kualitas
    quality_columns = data[:,-1]
    # Menghitung akurasi dari setiap class unik di kolom Species
    unique_classes, counts_unique_classes = np.unique(quality_columns, return_counts = True)
    
    # Cari index  yang paling banyak muncul(modus)
    index = counts_unique_classes.argmax()
    
    # Mengambil label klasifikasi dengan jumlah tertinggi
    classification = unique_classes[index]
    
    return classification
```

```python
def get_potential_splits(data):

    potential_splits = {}  # Inisialisasi kamus untuk menyimpan potensi pembagian
    _, n_columns = data.shape  # Mendapatkan jumlah baris dan kolom dari dataset

    # Iterasi melalui setiap kolom kecuali kolom terakhir (target variable)
    for column_index in range(n_columns - 1):
        potential_splits[column_index] = []  # Inisialisasi kamus untuk menyimpan potensi pembagian pada kolom tertentu
        values = data[:, column_index]  # Mengambil nilai-nilai pada kolom tersebut dari dataset
        unique_values = np.unique(values)

        for index in range(len(unique_values)):
            if index != 0:
                current_value = unique_values[index]
                previous_value = unique_values[index - 1]
                potential_split = (current_value + previous_value) / 2
                
                potential_splits[column_index].append(potential_split)

    return potential_splits
```

```python
def split_data(data, split_column, split_value):
    
    split_column_values = data[:, split_column]

    data_below = data[split_column_values <= split_value]
    data_above = data[split_column_values >  split_value]
    
    return data_below, data_above
```

```python
def calculate_entropy(data):
    
    label_column = data[:, -1]
    _, counts = np.unique(label_column, return_counts=True)

    probabilities = counts / counts.sum()
    entropy = sum(probabilities * -np.log2(probabilities))
     
    return entropy
```

```python
def calculate_overall_entropy(data_below, data_above):
    
    n = len(data_below) + len(data_above)
    p_data_below = len(data_below) / n
    p_data_above = len(data_above) / n

    overall_entropy =  (p_data_below * calculate_entropy(data_below) 
                      + p_data_above * calculate_entropy(data_above))
    
    return overall_entropy
```

```python
def determine_best_split(data, potential_splits):
    
    overall_entropy = 9999
    for column_index in potential_splits:
        for value in potential_splits[column_index]:
            data_below, data_above = split_data(data, split_column=column_index, split_value=value)
            current_overall_entropy = calculate_overall_entropy(data_below, data_above)

            if current_overall_entropy <= overall_entropy:
                overall_entropy = current_overall_entropy
                best_split_column = column_index
                best_split_value = value
    
    return best_split_column, best_split_value
```

```python
def decision_tree_algorithm(df, column_headers=None, counter=0, min_samples=2, max_depth=5):
    # data preparations
    if counter == 0:
        global COLUMN_HEADERS
        COLUMN_HEADERS = df.columns
        data = df.values
    else:
        data = df
    
    # base cases
    if (check_purity(data)) or (len(data) < min_samples) or (counter == max_depth):
        classification = classify_data(data)
        return classification

    # recursive part
    else:
        counter += 1

        # helper functions 
        potential_splits = get_potential_splits(data)
        split_column, split_value = determine_best_split(data, potential_splits)
        data_below, data_above = split_data(data, split_column, split_value)
        
        # instantiate sub-tree
        feature_name = column_headers[split_column]
        question = "{} <= {}".format(feature_name, split_value)
        sub_tree = {question: []}
        
        # find answers (recursion)
        yes_answer = decision_tree_algorithm(data_below, column_headers, counter, min_samples, max_depth)
        no_answer = decision_tree_algorithm(data_above, column_headers, counter, min_samples, max_depth)
        
        # If the answers are the same, then there is no point in asking the question.
        # This could happen when the data is classified even though it is not pure
        # yet (min_samples or max_depth base cases).
        if yes_answer == no_answer:
            sub_tree = yes_answer
        else:
            sub_tree[question].append(yes_answer)
            sub_tree[question].append(no_answer)
        
        return sub_tree
```

```python
def classify_example(example, tree):
    question = list(tree.keys())[0]

    # Split the question into parts based on spaces
    parts = question.split()

    # Determine the feature name and comparison operator
    feature_name = parts[0]
    comparison_operator = parts[1]

    # Join the rest of the parts to get the value (in case it contains spaces)
    value = ' '.join(parts[2:])

    try:
        # ask question
        if comparison_operator in ['<=', '<', '>', '>='] and len(parts) >= 3:
            example_value = example[feature_name]
            value = float(value) if '.' in value else int(value)  # Convert to float or int as needed
            if comparison_operator == "<=" and example_value <= value:
                answer = tree[question][0]
            elif comparison_operator == ">" and example_value > value:
                answer = tree[question][1]
            else:
                raise ValueError("Invalid comparison operator or condition not met")
        else:
            raise ValueError("Unexpected question structure")
    except ValueError as e:
        print(f"Error in question: {question}")
        print(f"Example value: {example_value}")
        print(f"Comparison operator: {comparison_operator}")
        print(f"Value in question: {value}")
        raise e

    # base case
    if not isinstance(answer, dict):
        return answer
    else:
        residual_tree = answer
        return classify_example(example, answer)
```

```python
def calculated_accuracy(test_df, tree):
    
    df["classification"] = df.apply(classify_example, axis = 1, args = (tree,))
    df["classification_correct"] = df.classification == df.quality
    accuracy = df.classification_correct.mean()
    
    return accuracy
```
