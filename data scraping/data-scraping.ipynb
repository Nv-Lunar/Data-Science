```python
import pandas as pd
import time

from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

- pergi ke web yang dituju contoh(link dibawah ini) lalu tekan f12

```python
# Siapkan url web yang akan di crawling
url = 'https://www.tokopedia.com/wecoolofficial/product'
```

- webdriver.Chrome(): Membuat instance dari objek WebDriver untuk browser Google Chrome. WebDriver digunakan untuk mengontrol browser secara otomatis.

- driver.get(url): Membuka URL yang diberikan pada browser yang dikendalikan oleh WebDriver. Fungsi ini digunakan untuk membuka halaman web yang sesuai dengan URL yang diberikan.

```python
driver = webdriver.Chrome()
driver.get(url)
```

- WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#zeus-root"))): Baris ini menggunakan WebDriverWait untuk menunggu hingga elemen dengan CSS selector #zeus-root muncul di halaman. Timeout maksimumnya adalah 5 detik. WebDriverWait digunakan untuk menangani situasi di mana suatu elemen membutuhkan waktu untuk dimuat atau muncul di halaman.

- for i in range(10): driver.execute_script("window.scrollBy(0,250)"): Baris ini adalah loop yang menjalankan JavaScript untuk melakukan scroll vertikal sebanyak 250 piksel setiap kali melalui iterasi loop. Dengan cara ini, halaman web akan di-scroll sebanyak 250 piksel sebanyak 10 kali.

```python
WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#zeus-root")))
for i in range(10):
    driver.execute_script("window.scrollBy(0,250)")
```

- Membuat objek BeautifulSoup yang digunakan untuk melakukan parsing HTML. driver.page_source digunakan untuk mendapatkan HTML dari halaman web yang telah diakses oleh WebDriver. "html.parser"

```python
soup = BeautifulSoup(driver.page_source, "html.parser")
print(soup)
```

```python
# Membuat column utuk dataframe karena nantinya data akan disimpan menggunakan metode loc pandas
data = pd.DataFrame(columns=['Product Name', 'Discount Price', 'Real Price', 'Discount', 'Sold'])
```

- findAll(tag html, class), digunakan untuk mengambil semua tag yang berada didalam tag dan class yang ditunjuk
- find(tag html, class), digunakan untuk mengmbil semua informasi yang berada di dalam tag dan class yang ditunjuk
- DataFrame.loc[rows, columns]

```python
# Perulangan untuk menemukan semua yang ada di dalam div dengan class = 'css-1asz3by' sebagai item
for item in soup.findAll('div', class_='css-1asz3by'):
    # Temukan di dalam item sebuah div dengan class = prd_link-product-name dan ambil teks nya
    product_name = item.find('div', class_='prd_link-product-name').text
    product_discount_price = item.find('div', class_='prd_link-product-price').text
    # Handle jika varibel memiliki data/tidak
    if len(product_discount_price) > 0:
        product_discount_price
    else:
        product_price = ''
    
    # set default variabel sebagai stringkosong untuk berjaga2 jika data tidak ada
    product_discount = ''  # Define product_discount here
    product_real_price = ''  # Define product_real_price here
    product_sold = ''  # Define product_sold here
    
    for item2 in item.findAll('div', class_='css-6cp8z8'):
        product_discount = item2.find('div', class_='prd_badge-product-discount').text
        product_real_price = item2.find('div', class_='prd_label-product-slash-price').text
        if len(product_discount) > 0:
            product_discount
        else:
            product_discount = ''
        if len(product_real_price) > 0:
            product_real_price
        else:
            product_real_price = ''
            
    for item in item.findAll('div', class_='css-yaxhi2'):
        for item2 in item:
            
            span_element = item2.find('span', class_='prd_label-integrity')
            product_sold = span_element.text if span_element else 'None'
            
    # Memasukan data pada data frame
    df = data.loc[len(data)] = {
        # pada column a masukkan data
        'Product Name': product_name,
        'Discount Price': product_discount_price,
        'Real Price': product_real_price,
        'Discount': product_discount,
        'Sold': product_sold
    }
    
    print('Product Name  :',product_name)
    print('Discount Price:',product_discount_price)
    print('Real Price    :',product_real_price)
    print('Discount      :',product_discount)
    print('Sold          :',product_sold)
    print('===='*15)
```
