```python
# Import pandas
import pandas as pd
```

```python
# Import csv
# sep=";" menjadikan semicolon(;) sebagai pembatas agar tabel rapi
data = pd.read_csv("Youtube_Data_Csv.csv", sep=";")

# Copy csv agar data asli tidak ter rewrite
data = data.copy()

# Menggunakan data frame agar data lebih terstruktur dan memudahkan pengembangan(Numoy, matplotlib, scikit-learn)
df = pd.DataFrame(data)

# Menampilkan semua data
# pd.set_option('display.max_rows', None)

# Reset pengaturan tampilan ke default
# pd.reset_option('display.max_rows')

df
```

```python
# Mengganti nama columns menjadi list
df = df.rename(columns = {'MAT': 'Matches_Played',
                          'NO' : 'Not_Outs',
                          'HS' : 'Highest_Inns_Score',
                          'BF' : 'Balls_Faced',
                          'SR' : 'Batting_Strike_Rate'})
# Menamilkan data paling atas/head
df.head()
```

```python
# Chek data kosong
df.isnull().any()
```

```python
# Menampilkan data yang Balls_Faced nya kosong(NaN)
df[df['Balls_Faced'].isna()==1]
```

```python
# Mengisi data kosong
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
```

```python
# Mengisi data kosong
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].fillna(0)
```

```python
# Membelah/memisahkan Span menjadi Start dan End Date dengan - sebagai pembatas
df['Span'].str.split(pat = '-')
```

```python
# Membelah/memisahkan Span dengan - sebagai pembatas dan menampilkan data index 0
df['Span'].str.split(pat = '-').str[0]
```

```python
# Membuat kolom baru dengan nama Rookie_Year dan data index 0 dari hasil Split span
# Mirip alter table di sql
df['Rookie_Year'] = df['Span'].str.split(pat = '-').str[0]
```

```python
# Membuat kolom baru dengan nama Final_Year dan data index 1 dari hasil Split span
df['Final_Year'] = df['Span'].str.split(pat = '-').str[1]
```

```python
# Menghapus kolom span
# axis = 1 berarti menhapus menghapus kolom
# axis = 1 berarti menhapus menghapus baris 
df = df.drop(['Span'], axis = 1)
```

```python
# Membelah negara dari Player
df['Player'].str.split(pat = '(')
```

```python
df['Country'] = df['Player'].str.split(pat = '(').str[1]
```

```python
df['Country'] = df['Country'].str.split(pat = ')').str[0]
```

```python
# Mengambil nama player yang mana ada di index[0] dengan ( sebagai pemisah
df['Player'] = df['Player'].str.split(pat = '(').str[0]
```

```python
# Mengubah DataTypes
df.dtypes
```

```python
df['Highest_Inns_Score'].str.split(pat = '*').str[0]
```

```python
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split(pat = '*').str[0]
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype('int')
```

```python
df['Balls_Faced'].str.split(pat = '+').str[0]
```

```python
df['Balls_Faced'] = df['Balls_Faced'].str.replace('-','+').str.split(pat = '+').str[0]

# .replace(a,b) mengubah data a menjadi b
# df['Balls_Faced'] = df['Balls_Faced'].replace('', '0')
# pd.reset_option('display.max_rows')
df['Balls_Faced']
```

```python
# pd.to_numeric adalah fungsi panda yang mengubah argumen menjadi tipe numerik. 
# 'coerce': Memaksa nilai yang tidak valid menjadi NaN(Bukan Angka) alih-alih memunculkan kesalahan.
# mengubah dtype menjadi int dengan astype('int')
df['Balls_Faced'] = pd.to_numeric(df['Balls_Faced'], errors='coerce').fillna(0).astype(int)

df['Balls_Faced']
```

```python
# Mengubah ke float
df['Ave'] = df['Ave'].astype(float)
# Mengubah ke float dan menyingkirkan elemen menggangu
df['Batting_Strike_Rate'] = pd.to_numeric(df['Batting_Strike_Rate'], errors='coerce').fillna(0).astype(float)
df.dtypes
```

```python
df = df.astype({'Rookie_Year': 'int', 'Final_Year' : 'int'})
```

```python
# Membuat kolom Career_Year dengan akhir taun kerja - awal tahun kerja
df['Career_Length']= df['Final_Year'] - df['Rookie_Year']
```
