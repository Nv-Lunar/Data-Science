```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
iris = load_iris()
data = iris.data
target = iris.target
```

```python
target
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.3)
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
```

```python
accuracy_score(y_test, predictions)
```

```python
knn = KNeighborsClassifier()
```

```python
param_grid = {"n_neighbors" : range(1,30)}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(knn, param_grid, cv = 2)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_k = grid_search.best_params_["n_neighbors"]
```

```python
best_knn = KNeighborsClassifier(n_neighbors = best_k)
```

```python
best_knn.fit(X_train, y_train)
```

```python
predictions = best_knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
```

```python
accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()
```

```python
dt_classifier.fit(X_train, y_train)
```

```python
predictions = dt_classifier.predict(X_test)
```

```python
accuracy_score(y_test, predictions)
```

```python
param_grid = {
    "max_depth" : range(1,11),
    "criterion" : ["gini", "entropy"]
}
```

```python
grid_search = GridSearchCV(dt_classifier, param_grid, cv = 3)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_depth = grid_search.best_params_["max_depth"]
```

```python
best_criterion = grid_search.best_params_["criterion"]
```

```python
best_dt_classifier = DecisionTreeClassifier(max_depth = best_depth, criterion = best_criterion)
```

```python
best_dt_classifier.fit(X_train, y_train)
```

```python
predictions = best_dt_classifier.predict(X_test)
```

```python
accuracy_score(y_test, predictions)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
```

```python
diabetes = load_diabetes()
data = diabetes.data
target = diabetes.target
```

```python
target
```

```python
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.2)
```

```python
linear_reg_model = LinearRegression()
```

```python
linear_reg_model.fit(X_train, y_train)
```

```python
predictions = linear_reg_model.predict(X_test)
```

```python
predictions
```

```python
mean_squared_error(y_test, predictions)
```

```python
r2_score(y_test, predictions)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge()
```

```python
ridge_model.fit(X_train, y_train)
```

```python
predictions = ridge_model.predict(X_test)
```

```python
mean_squared_error(y_test, predictions)
```

```python
r2_score(y_test, predictions)
```

```python
param_grid = {
    "alpha" : [0.1, 1, 10, 100]
}
```

```python
grid_search = GridSearchCV(ridge_model, param_grid, cv = 2)
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_alpha = grid_search.best_params_["alpha"]
```

```python
best_ridge_model = Ridge(alpha = best_alpha)
```

```python
best_ridge_model.fit(X_train, y_train)
```

```python
predictions = best_ridge_model.predict(X_test)
predictions_train = best_ridge_model.predict(X_train)
```

```python
print("train :", mean_squared_error(y_train, predictions_train))
print("test :", mean_squared_error(y_test, predictions))
```

```python
print("train :", r2_score(y_train, predictions_train))
print("test  :", r2_score(y_test, predictions))
```
